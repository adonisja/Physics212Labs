```python
display(Math(r'''
\text{Temperature in gases is a measure of the average translational kinetic energy of particles:}
'''))

display(Math(r'''
T \propto \left\langle \frac{1}{2}mv^2\right\rangle
'''))

display(Math(r'''
\text{Where:} \\
\quad T = \text{Absolute temperature (Kelvin)} \\
\quad m = \text{Mass of one gas particle} \\
\quad \overline{v^2} = \text{Mean square velocity of particles}
'''))

# Key relationships
display(Math(r'''
\text{Key relationships:} \\
1.\ \text{As } T \text{ increases, particle speeds increase} \\
2.\ v_{rms} = \sqrt{\frac{3k_BT}{m}} \\
3.\ \text{Maxwell-Boltzmann distribution describes velocity spread}
'''))

# Example calculation
def rms_speed(T, molar_mass):
    """Calculate RMS speed in m/s"""
    R = 8.314  # J/(mol·K)
    return (3 * R * T / molar_mass)**0.5

display(Math(rf'''
\text{{Example}}: N_2 \text{{ molecules at 298 K: }} {rms_speed(298, 0.028):.0f} \text{{ m/s}}
'''))
```

```python
display(Math(r'''
\text{Charles' Law for absolute zero determination:} \\
\frac{V}{T} = k \quad \text{or} \quad T = \frac{V}{k} - 273.15°C
'''))

display(Math(r'''
\text{With } \pm1°C \text{ systematic error:} \\
T_{\text{measured}} = T_{\text{actual}} + 1
'''))

# Effect on absolute zero determination
display(Math(r'''
\text{Determining absolute zero involves extrapolating } V \rightarrow 0: \\
T_{\text{abs 0 (measured)}} = T_{\text{abs 0 (actual)}} + 1°C \\
\boxed{\text{A }+1°C\text{ error shifts your absolute zero determination by }+1°C}
'''))

# Practical example
actual_temps = [100, 50, 25, 0, -25, -50]  # °C
measured_temps = [t+1 for t in actual_temps]

display(Math(r'''
\text{Example:} \\
\begin{array}{|c|c|}
\hline
\text{Actual Temp (°C)} & \text{Measured Temp (°C)} \\
\hline
100 & 101 \\
50 & 51 \\
25 & 26 \\
0 & 1 \\
-25 & -24 \\
-50 & -49 \\
\hline
\end{array} \\
\text{Linear fit will intersect "zero volume" at } -272.15°C \text{ instead of } -273.15°C
'''))
```

```python
T_true = np.array([100, 50, 25, 0, -25, -50, -273.15])  
V = (T_true + 273.15) * 0.1  

# With +1°C error
T_measured = T_true + 1

plt.figure(figsize=(10,5))
plt.plot(T_true, V, 'b-', label='True relationship')
plt.plot(T_measured, V, 'r--', label='With +1°C error')
plt.axvline(-273.15, color='blue', linestyle=':', label='True absolute zero')
plt.axvline(-272.15, color='red', linestyle=':', label='Measured absolute zero')
plt.xlabel('Temperature (°C)')
plt.ylabel('Volume (arb. units)')
plt.legend()
plt.grid()
plt.show()
```

This systematic error doesn't affect the slope of the line (so gas law calculations remain internally consistent), but it directly offsets the absolute zero determination by the same +1°C error.

```python
display(Markdown("### Gay-Lussac’s Law and Absolute Zero Error Analysis"))

# Equation - Gay-Lussac’s Law
display(Math(r"\text{Gay-Lussac's Law:} \quad \frac{P}{T} = k \quad \Rightarrow \quad P = k(T + 273.15)"))

# Systematic Error Introduction
display(Math(r"\text{With a } +0.10 \,\text{atm error:}"))
display(Math(r"P_{\text{measured}} = P_{\text{true}} + 0.10 = k(T + 273.15) + 0.10"))

# Solving for Absolute Zero
display(Math(r"\text{At absolute zero } (P = 0):"))
display(Math(r"0 = k(T_{\text{abs}} + 273.15) + 0.10"))
display(Math(r"T_{\text{abs}} = -273.15 - \frac{0.10}{k}"))

# Error Magnitude Definition
display(Math(r"\Delta T = -\frac{0.10}{k} \quad \text{where} \quad k = \frac{P_{\text{true}}}{T_{\text{true}}}"))

# Plug in an example
display(Math(r"\text{Example: } T = 298\,\text{K}, \quad P = 1.00\,\text{atm}"))
display(Math(r"k = \frac{1.00}{298} \approx 0.00336\,\text{atm/K}"))
display(Math(r"\Delta T = -\frac{0.10}{0.00336} \approx -29.8\,\text{K}"))
display(Math(r"T_{\text{abs}} = -273.15 - 29.8 = -302.95\,^\circ\text{C}"))

# Final Summary
display(Math(r"\boxed{\text{A } +0.10\,\text{atm systematic error causes } \approx -30^\circ\text{C error in absolute zero}}"))
```

```python
# Convert temperatures to Kelvin
df_clean['Temperature (K)'] = df_clean['Temperature (°C)'] + 273.15

# Linear regression analysis
def analyze_run(run_label, color):
    # Filter for specific run
    run_data = df_clean[df_clean['Run'] == run_label]
    X = run_data[['Temperature (K)']]
    y = run_data['Pressure (kPa)']
    
    # Fit linear model
    model = LinearRegression()
    model.fit(X, y)
    y_fit = model.predict(X)
    
    # Plot data and best fit line
    plt.figure(figsize=(10, 6))
    plt.scatter(X, y, label=f'{run_label} Data', color=color)
    plt.plot(X, y_fit, color=color, label=f'Best fit: P = {model.intercept_:.2f} + {model.coef_[0]:.2f}T')
    plt.xlabel('Temperature (K)')
    plt.ylabel('Pressure (kPa)')
    plt.title(f'Pressure vs. Temperature for {run_label}')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    # Residuals analysis
    residuals = y - y_fit
    plt.figure(figsize=(10, 4))
    plt.scatter(X, residuals, color=color, alpha=0.7)
    plt.axhline(0, color='black', linestyle='--')
    plt.title(f'Residuals for {run_label}')
    plt.xlabel('Temperature (K)')
    plt.ylabel('Residuals (kPa)')
    plt.grid(True)
    plt.show()
    
    # Print regression stats
    r_squared = model.score(X, y)
    y_intercept = model.intercept_
    slope = model.coef_[0]
    T_absolute_zero_K = -y_intercept / slope
    T_absolute_zero_C = T_absolute_zero_K - 273.15
    percent_error = abs(T_absolute_zero_C + 273.15) / 273.15 * 100
    
    print(f"📍 {run_label} Experimental absolute zero (K): {T_absolute_zero_K:.2f} K")
    print(f"📉 {run_label} Experimental absolute zero (°C): {T_absolute_zero_C:.2f} °C")
    print(f"🔹 {run_label} Percent Error: {percent_error:.2f}%")
    print(f"📊 {run_label} R²: {r_squared:.4f}\n")
    
    return r_squared

# Analyze Run B and Run C
r2_run_b = analyze_run("Run B", color="red")
r2_run_c = analyze_run("Run C", color="blue")

# Analysis
print("🔎 **Analysis:**")
if r2_run_b > 0.9 and r2_run_c > 0.9:
    print("Both Run B and Run C exhibit strong linear relationships, as indicated by high R² values. This suggests that the pressure-temperature relationship is approximately linear, consistent with Gay-Lussac's law for ideal gases at moderate pressures.")
else:
    print("At least one of the runs shows a weak linear relationship, indicating potential experimental errors or non-ideal gas behavior.")
```

### ❓ Why Do My Experimental Values of Absolute Zero Differ from the Theoretical Value?

⚠️ Our experimental values for absolute zero differ from the theoretical value of **−273.15°C** due to several possible reasons:

1. **Instrumental Errors**  
   Measurement devices like thermometers or pressure sensors may be miscalibrated, introducing consistent bias.

2. **Non-Ideal Gas Behavior**  
   Gases deviate from ideal behavior at very low temperatures due to intermolecular forces, violating assumptions of Gay-Lussac’s Law.

3. **Limited Data Range**  
   Our data only extends down to around −72°C, so the estimation of absolute zero requires **extrapolation far beyond the measured values**, which increases uncertainty.

4. **Human or Recording Error**  
   Mistakes during data entry, timing inconsistencies, or unit conversion issues can impact the accuracy of the regression line.

5. **System Leaks or Volume Changes**  
   If the gas container was not perfectly sealed or changed volume with temperature, this would invalidate the assumption of constant volume.

6. **Environmental Influences**  
   Factors like ambient pressure, temperature drift, or vibrations could affect pressure readings.

> 💡 These sources of error contribute to the deviation of our experimental estimate from the true value.

### 🎈 Balloon and Liquid Nitrogen Demonstration: Observations and Explanation

#### 🔬 Observation:
When a **balloon filled with air** is placed into **liquid nitrogen (−196°C)**:
- The balloon **shrinks rapidly** and becomes shriveled.
- After removing it from the liquid nitrogen and allowing it to warm up, the balloon **expands back to its original size** (or close to it).

#### 🧠 Explanation:

1. **Gas Particle Kinetic Energy Drops**  
   Cooling the balloon causes the **air molecules inside to lose kinetic energy**, move more slowly, and collide with the balloon walls less frequently and with less force.

2. **Pressure Decreases**  
   According to **Gay-Lussac’s Law** (\( P \propto T \)), when the temperature of a gas decreases (in Kelvin), its pressure also decreases if the volume is constant. But since the balloon is flexible, the **volume also decreases** in response to this pressure drop.

3. **Balloon Shrinks Due to Lower Internal Pressure**  
   The internal pressure becomes much lower than the surrounding atmospheric pressure, so the balloon **collapses inward**.

4. **Re-Expansion Upon Warming**  
   As the air inside warms up, kinetic energy increases, restoring pressure and volume — the balloon **re-inflates** as it approaches room temperature.

#### ✅ Conclusion:
This demonstration visually confirms that **temperature, pressure, and volume are directly related** in gases. Cooling reduces molecular motion, which lowers pressure and volume, consistent with the **ideal gas law** and **Gay-Lussac’s Law**.


### ⚠️ Sources of Error in the Experimental Setup

During the setup and execution of the experiment to estimate absolute zero, several potential **sources of error** may arise:

---

#### 1. **Thermometer Calibration and Placement**
- If the thermometer is **miscalibrated** or not fully immersed in the gas, the measured temperature may not reflect the true temperature of the gas.
- Lag in temperature response can also cause inaccurate readings.

#### 2. **Pressure Sensor Accuracy**
- A **poorly calibrated pressure sensor** or one with limited resolution can lead to incorrect pressure values.
- Fluctuations due to vibrations or ambient air drafts can introduce noise.

#### 3. **Gas Leakage**
- Small **leaks in the tubing or container** can cause loss of gas or pressure, skewing the pressure–temperature relationship.

#### 4. **Non-Isolated System**
- If the system is not well insulated, **external temperature changes** (e.g., room drafts, hand warmth) can influence readings.

#### 5. **Non-Ideal Gas Effects**
- At low temperatures or high pressures, the gas may **deviate from ideal behavior**, violating assumptions of the ideal gas law.

#### 6. **Volume Changes**
- If the gas container expands/contracts with temperature, it affects pressure in a way that invalidates the assumption of **constant volume**.

#### 7. **Human Error**
- Misreading scales, recording incorrect values, or using inconsistent timing during measurements can introduce random or systematic error.

---

> ✅ Identifying these sources helps explain why the extrapolated absolute zero might differ from the theoretical value and allows for better refinement in future experiments.
